```python
df = pd.read_excel('data.xlsx')
df2 = pd.read_excel('data.xlsx', sheet_name='Networks')
```

## Fill out the network name in the Dec Spend sheet

```python
ids = []
for i in df['Source ID']:
    x = df2.loc[df2['Network id'] == i, 'Company Name'].iloc[0]
    ids.append(x)
        
df['Network Name'] = ids
```

```python
df
```

## Find the spend for each network

```python
spend = []

for i in df2['Company Name']:
    network = df.loc[df['Network Name'] == i]
    x = network.Spend.sum()
    spend.append(x)
```

```python
df2['spend per network'] = spend
df2
```

## Find the spend for each domain

```python
domains = pd.DataFrame(data=set(list(df['Domain or App'])), columns=['domain'])
```

```python
spending = []

for i in domains.domain:
    domain_filter = df.loc[df['Domain or App'] == i]
    x = domain_filter.Spend.sum()
    spending.append(x)
    
domains['spend per domain'] = spending
```
